```python
def find_largest(number_list:int) -> int:
    if not isinstance(number_list,int):
        raise TypeError("input must be a int")
    if len(number_list) == 0:
        raise ValueError("list is empty")
    largest = number_list[0]
    for num in list1:
        if num >largest:
           largest = num
    return largest  
# main -- excuetion
if __name__ == "__main__":
    try:
        # taking inout from the user
        user_input = input("enter a number seperate by comma")
        # convert into int
        number_list = [int(item) for item in user_input.split(",")]
        #calling a function
        result = find_largest(number_list)
        print(f"largest number is {result}")
    except (TypeError,ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
```
